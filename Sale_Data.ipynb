#ขั้นตอนที่ 1: Setup & Data Loading (การตั้งค่าและนำเข้าข้อมูล)
- เราจะเชื่อมต่อ Google Colab เข้ากับ Google Drive

- เขียนโค้ดให้ Python วนลูปอ่านไฟล์ยอดขายรายเดือนทุกไฟล์ แล้วนำมาต่อกัน (Concatenate) ให้เป็นตาราง (DataFrame) ก้อนเดียวที่สมบูรณ์

```python
# 1. ขอสิทธิ์การเข้าถึง google drive
from google.colab import drive
drive.mount('/content/drive')
```

```python
# 2. นำเข้า library
import pandas as pd
import glob
import os

# 3. กำหนดที่อยู่ของโฟลเดอร์ที่คุณเก็บไฟล์ยอดขายไว้
folder_path = '/content/drive/MyDrive/12. Banyasookjai/revenue'

# 4. ใช้คำสั่ง glob เพื่อค้นหาไฟล์ทุกไฟล์ที่นามสกุล .xls หรือ .xlsx ในโฟลเดอร์นี้
all_files = glob.glob(os.path.join(folder_path,"*xls*"))
print(f"ค้นพบไฟล์ทั้งหมด:{len(all_files)} เดือน")

# 5.สร้าง list [] ว่างๆมารอรับข้อมูล
df_list = []

# 6. loop เปิดอ่านไฟล์ทีละเดือน แล้วนำมาเก็บไว้ใน List
for file in all_files:
  print(f"กำลังอ่านไฟล์:{os.path.basename(file)}")
  # อ่านไฟล์ Excle โดยใช้ pandas
  df_temp = pd.read_excel(file)
  df_list.append(df_temp)

# 7. นำข้อมูลของทุกเดือนมาต่อกัน(concatenate) ให้เป็นตารางเดียว
if len(df_list) > 0:
  df = pd.concat(df_list, ignore_index=True)
  print("รวมไฟล์สำเร็จ")
  print(f"รายการขายทั้งหมดตั้งแต่เดือนแรกจนถึงเดือนล่าสุด:{df.shape[0]}รายการ")
else:
  print("ไม่พบไฟล์ใน folder")
```

* pandas คือไลบรารีที่ใช้จัดการข้อมูลในรูปแบบตาราง

เราใช้ for loop วิ่งเข้าไปอ่านทีละไฟล์ การเขียนแบบนี้ทำให้เดือนหน้าพอมีไฟล์ใหม่เข้ามาใส่ในโฟลเดอร์ โค้ดนี้ก็จะกวาดมาอ่านให้ทั้งหมดโดยที่ไม่ต้องแก้โค้ดเลย

#สเต็ปที่ 2: การทำความสะอาดและเตรียมข้อมูล (Data Cleaning & Preprocessing)

```python
# 1. สร้างสำเนาข้อมูล
df_clean = df.copy() #เป็นการก๊อปปี้ตารางข้อมูลเดิม (df) มาสร้างเป็นตารางใหม่ชื่อ df_clean

# 2. ลบแถวที่ไม่มี "เลขที่ใบเสร็จรับเงิน"
df_clean = df_clean.dropna(subset=['เลขที่ใบเสร็จรับเงิน'])

#2.1 โค้ดที่แทรกใหม่: ลบรายการที่เป็น "ค่าบริการ CI" ออกจากการวิเคราะห์
df_clean = df_clean[df_clean['รายการสินค้า'] != 'ค่าบริการ CI']

# 3. จัดการ column วันที่ขาย
if pd.api.types.is_numeric_dtype(df_clean['วันที่ขาย']): #คำสั่ง Drop NaN (Not a Number) คือการ "ลบแถวที่มีค่าว่าง" โดยระบุคอลัมน์ที่เจาะจงผ่าน subset
  df_clean['วันที่ขาย'] = pd.to_datetime(df_clean['วันที่ขาย'], origin='1899-12-30', unit='D')
else:
    df_clean['วันที่ขาย'] = pd.to_datetime(df_clean['วันที่ขาย']) #คำสั่งแปลงข้อมูลให้กลายเป็นรูปแบบ วัน/เดือน/ปี (Datetime)

# 4. สร้าง column ใหม่ สำหรับ เดือน และปี
# เป็นคำสั่งสกัดเอาเฉพาะ "เลขเดือน" และ "เลขปี ค.ศ." ออกมาจากคอลัมน์วันที่
df_clean['เดือน'] = df_clean['วันที่ขาย'].dt.month
df_clean['ปี'] = df_clean['วันที่ขาย'].dt.year

# 5. แปลงคอลัมน์ตัวเลขให้พร้อมคำนวณ
# บังคับให้คอลัมน์เกี่ยวกับการเงินและจำนวน กลายเป็นประเภท "ตัวเลข" ที่บวกลบคูณหารได้
numeric_columns = ['จำนวน','ราคาต้นทุนต่อหน่วย','ราคาขายต่อหน่วย','ส่วนลดรวม','ราคาขายรวม','กำไรรวม']
for col in numeric_columns:
  df_clean[col] = pd.to_numeric(df_clean[col], errors='coerce') # (คำสั่ง errors='coerce' จะช่วยจัดการแถวที่มีปัญหาให้กลายเป็นค่าว่าง แทนที่จะเกิด Error)

# 6. ลบแถวที่ตัวเลขผิดพลาดจนกลายเป็นค่าวาง
df_clean = df_clean.dropna(subset=['กำไรรวม','ราคาขายรวม'])

print("ทำความสะอาดข้อมูลเสร็จแล้ว")
print(f"จำนวนรายการที่พร้อมวิเคราะห์จริง:{df_clean.shape[0]} รายการ")

# แสดงตัวอย่าง 5 บรรทัดแรก
display(df_clean[['เลขที่ใบเสร็จรับเงิน', 'วันที่ขาย', 'รายการสินค้า', 'จำนวน', 'กำไรรวม']].head())
```

```python
# 1. บังคับแปลงคอลัมน์ 'วันที่ขาย' ให้เป็น Datetime ก่อน
# (คำสั่ง errors='coerce' จะช่วยจัดการแถวที่มีปัญหาให้กลายเป็นค่าว่าง แทนที่จะเกิด Error)
df_clean['วันที่ขาย'] = pd.to_datetime(df_clean['วันที่ขาย'], errors='coerce')

# 2. เมื่อมั่นใจว่าเป็น Datetime แล้ว ค่อยใช้ .dt.date หั่นเอาเฉพาะวันที่ (ทิ้งเวลา)
df_clean['วันที่ขาย'] = df_clean['วันที่ขาย'].dt.date

# 3. แสดงผลลัพธ์ 5 บรรทัดแรก เพื่อเช็คความเรียบร้อย
print("✅ แปลงวันที่สำเร็จแล้ว!")
display(df_clean[['เลขที่ใบเสร็จรับเงิน', 'วันที่ขาย', 'เวลา']].head())
```

#ขั้นตอนที่ 3: การวิเคราะห์ข้อมูลเบื้องต้น (Exploratory Data Analysis - EDA)

```python
# 1. ดูภาพรวมยอดขายและกำไรทั้งหมด
total_revenue = df_clean['ราคาขายรวม'].sum()
total_profit = df_clean['กำไรรวม'].sum()

print("📊 ภาพรวมธุรกิจ (Overall Performance):")
# ใช้ :,.2f เพื่อใส่ลูกน้ำและทศนิยม 2 ตำแหน่งให้ดูง่าย
print(f"ยอดขายรวมทั้งหมด: {total_revenue:,.2f} บาท")
print(f"กำไรรวมทั้งหมด: {total_profit:,.2f} บาท")
print("-" * 50)

# 2. หา Top 10 สินค้าที่สร้างกำไรสูงสุด (สินค้าหัวใจหลักของร้าน)
top_profit = df_clean.groupby('รายการสินค้า').agg({
    'กำไรรวม': 'sum',
    'จำนวน': 'sum'
}).reset_index()
top_10_profit = top_profit.sort_values(by='กำไรรวม', ascending=False).head(10)

print("\n🏆 10 อันดับสินค้าที่ทำกำไรสูงสุด:")
display(top_10_profit)

# 3. หา Top 10 สินค้าที่ขายออกบ่อยที่สุด (วัดจากจำนวนชิ้น)
top_10_volume = top_profit.sort_values(by='จำนวน', ascending=False).head(10)

print("\n📦 10 อันดับสินค้าที่ขายดีที่สุด (นับตามจำนวนชิ้น):")
display(top_10_volume)

# 4. ดูยอดขายรายเดือน (เพื่อดูความเติบโตและฤดูกาล)
df_clean['ปี-เดือน'] = pd.to_datetime(df_clean['วันที่ขาย']).dt.to_period('M')
monthly_sales = df_clean.groupby('ปี-เดือน').agg({  # aggregate
    'ราคาขายรวม': 'sum',
    'กำไรรวม': 'sum'
}).reset_index()

print("ยอดขายและกำไร แยกตามเดือน:")
display(monthly_sales)

# ---------------------------------------------------------
# 💾 ส่วนบันทึกไฟล์: เซฟทั้ง 3 ตารางลงในไฟล์ Excel เดียวกัน (แต่แยก Sheet)
# ---------------------------------------------------------
print("\n💾 กำลังเซฟไฟล์ลง Google Drive...")

# กำหนดชื่อไฟล์หลักที่จะเซฟ
path_to_save = '/content/drive/MyDrive/12. Banyasookjai/revenue/สรุปภาพรวมยอดขาย_EDA.xlsx'

# ใช้คำสั่ง ExcelWriter เพื่อเปิดไฟล์ขึ้นมาเขียนหลายๆ หน้า
with pd.ExcelWriter(path_to_save) as writer:
    # เซฟตารางที่ 1 ลง Sheet ชื่อ 'Top10_กำไรสูงสุด'
    top_10_profit.to_excel(writer, sheet_name='Top10_กำไรสูงสุด', index=False)

    # เซฟตารางที่ 2 ลง Sheet ชื่อ 'Top10_ขายดีสุด'
    top_10_volume.to_excel(writer, sheet_name='Top10_ขายดีสุด', index=False)

    # เซฟตารางที่ 3 ลง Sheet ชื่อ 'ยอดขายรายเดือน'
    monthly_sales.to_excel(writer, sheet_name='ยอดขายรายเดือน', index=False)

print("✅ เซฟไฟล์ Excel สำเร็จแล้ว! ทั้ง 3 ตารางถูกรวมอยู่ในไฟล์เดียว (แยกแท็บด้านล่าง) ไปเปิดดูในโฟลเดอร์ revenue ได้เลยครับ")
```

#สเต็ปที่ 5 (สเต็ปสุดท้าย: Market Basket Analysis หาว่ายาอะไรที่ลูกค้าซื้อคู่กัน)

#Insight ข้อที่ 5: วิเคราะห์พฤติกรรมขนาดตะกร้า (Basket Size & Value Distribution)

```python
# ---------------------------------------------------------
# 🛒 Insight 5: วิเคราะห์ขนาดตะกร้า (Basket Size & Ticket Size)
# ---------------------------------------------------------
import matplotlib.pyplot as plt
import seaborn as sns

# 1. จัดกลุ่มตาม 'เลขที่ใบเสร็จรับเงิน' เพื่อรวมยอดขายของแต่ละบิล
basket_summary = df_clean.groupby('เลขที่ใบเสร็จรับเงิน').agg({
    'ราคาขายรวม': 'sum', # ยอดเงินรวมต่อบิล
    'จำนวน': 'sum'       # จำนวนชิ้นรวมต่อบิล
}).reset_index()

# 2. คำนวณหาค่าเฉลี่ย
avg_ticket_size = basket_summary['ราคาขายรวม'].mean()
avg_basket_size = basket_summary['จำนวน'].mean()

print(f"💰 ยอดใช้จ่ายเฉลี่ยต่อ 1 ใบเสร็จ (Ticket Size): {avg_ticket_size:,.2f} บาท")
print(f"🛍️ จำนวนสินค้าเฉลี่ยต่อ 1 ใบเสร็จ (Basket Size): {avg_basket_size:,.2f} ชิ้น\n")

# 3. วาดกราฟการกระจายตัว (Histogram) เพื่อดูช่วงราคาที่ลูกค้านิยมจ่าย
plt.figure(figsize=(10, 6))

# กรองเอาเฉพาะบิลที่ยอดไม่เกิน 1,500 บาท เพื่อให้กราฟดูง่าย (ตัดบิลที่แพงผิดปกติออกไปก่อน)
normal_bills = basket_summary[basket_summary['ราคาขายรวม'] <= 1500]

# วาด Histogram
sns.histplot(data=normal_bills, x='ราคาขายรวม', bins=30, color='purple', kde=True)

# ตกแต่งกราฟ
plt.title('🛒 การกระจายตัวของยอดใช้จ่ายต่อบิล (เฉพาะบิลไม่เกิน 1,500 บาท)')
plt.xlabel('ยอดใช้จ่ายต่อบิล (บาท)')
plt.ylabel('จำนวนบิล (ใบเสร็จ)')

# ขีดเส้นประสีแดง เพื่อแสดงจุดที่เป็น "ค่าเฉลี่ย"
plt.axvline(avg_ticket_size, color='red', linestyle='dashed', linewidth=2, label=f'ค่าเฉลี่ย {avg_ticket_size:.2f} บาท')
plt.legend()
plt.show()
```

#Insight ข้อที่ 4: วิเคราะห์สินค้าค้างสต็อกหรือเคลื่อนไหวช้า (Dead Stock / Slow Moving)

```python
# ---------------------------------------------------------
# 📦 Insight 4: วิเคราะห์สินค้าเคลื่อนไหวช้า / ค้างสต็อก (Slow Moving)
# ---------------------------------------------------------

# 1. จัดกลุ่มสินค้า และรวม 'จำนวน' ชิ้นที่ขายได้ทั้งหมดตลอดระยะเวลาที่มีข้อมูล
product_sales = df_clean.groupby('รายการสินค้า')['จำนวน'].sum().reset_index()

# 2. เรียงลำดับจากสินค้าที่ขายได้ "น้อยที่สุด" ไปหา "มากที่สุด" (ascending=True)
slow_moving = product_sales.sort_values(by='จำนวน', ascending=True)

# 3. กำหนดเกณฑ์ (Threshold) ว่าขายได้น้อยกว่ากี่ชิ้นถึงจะเรียกว่า "ขายยาก/ค้างสต็อก"
# (ในที่นี้ตั้งไว้ที่ 5 ชิ้น คุณแมนสามารถแก้เลข 5 เป็นเลขอื่นได้ตามความเหมาะสมของร้านครับ)
threshold = 5
dead_stock = slow_moving[slow_moving['จำนวน'] <= threshold]

print(f"⚠️ พบสินค้าที่เข้าข่ายค้างสต็อก (ขายได้ไม่เกิน {threshold} ชิ้น): {dead_stock.shape[0]} รายการ\n")

# 4. แสดงตัวอย่างสินค้าค้างสต็อก 15 อันดับแรกที่ขายได้น้อยที่สุด
print("🔻 ตัวอย่างรายการสินค้าที่ขายออกยากที่สุด (เรียงจากน้อยไปมาก):")
display(dead_stock.head(15))

# เซฟเป็นไฟล์ Excel ไปไว้ในโฟลเดอร์ revenue ใน Drive
path_to_save = '/content/drive/MyDrive/12. Banyasookjai/รายการสินค้าที่ขายออกยากที่สุด.xlsx'

# คำสั่งเซฟ (index=False คือไม่เอาตัวเลขลำดับแถว 0,1,2,3... ติดไปด้วย)
dead_stock.to_excel(path_to_save, index=False)
print("✅ เซฟไฟล์ Excel สำเร็จแล้ว ไปดูใน Drive ได้เลย!")
```

#Insight ข้อที่ 8: วิเคราะห์สัดส่วนกำไร (Margin Spread Analysis)

```python
# ---------------------------------------------------------
# 💎 Insight 8: วิเคราะห์สัดส่วนกำไร (% Margin Spread Analysis)
# ---------------------------------------------------------
import matplotlib.pyplot as plt
import seaborn as sns

# 1. จัดกลุ่มตามสินค้า รวมยอดขาย กำไร และจำนวนชิ้น
margin_df = df_clean.groupby('รายการสินค้า').agg({
    'ราคาขายรวม': 'sum',
    'กำไรรวม': 'sum',
    'จำนวน': 'sum'
}).reset_index()

# 2. กรองข้อมูล (Filter) เอาเฉพาะสินค้าที่ขายได้อย่างน้อย 20 ชิ้น
# (เพื่อป้องกันสินค้าที่นานๆ ขายได้ทีแต่ % กำไรสูง มาหลอกตาเรา)
margin_df = margin_df[margin_df['จำนวน'] >= 20]

# 3. คำนวณ % กำไรขั้นต้น (Gross Profit Margin)
# สูตรคือ: (กำไรรวม / ราคาขายรวม) * 100
margin_df['% กำไรขั้นต้น'] = (margin_df['กำไรรวม'] / margin_df['ราคาขายรวม']) * 100

# 4. เรียงลำดับหาสินค้าที่มี % กำไรสูงสุด 10 อันดับแรก
top_margin_products = margin_df.sort_values(by='% กำไรขั้นต้น', ascending=False).head(10)

print("💎 10 อันดับสินค้าที่มี % กำไรขั้นต้นสูงที่สุด (และมียอดขายเกิน 20 ชิ้น):")
# กำหนดให้โชว์แค่คอลัมน์ที่จำเป็น เพื่อให้ดูง่าย
display(top_margin_products[['รายการสินค้า', 'ราคาขายรวม', 'กำไรรวม', '% กำไรขั้นต้น']])

# 5. วาดกราฟเพื่อให้เห็นภาพชัดเจน
plt.figure(figsize=(12, 6))
# วาดกราฟแท่งแนวนอน
sns.barplot(data=top_margin_products, x='% กำไรขั้นต้น', y='รายการสินค้า', palette='magma')

plt.title('💎 10 อันดับสินค้าที่มี % กำไร (Margin) สูงที่สุด')
plt.xlabel('% กำไรขั้นต้น (เปอร์เซ็นต์)')
plt.ylabel('ชื่อสินค้า')

# 6. เพิ่มตัวเลข % ไว้ที่ปลายกราฟแท่งแต่ละอัน
for index, value in enumerate(top_margin_products['% กำไรขั้นต้น']):
    plt.text(value + 0.5, index, f'{value:.1f}%', va='center')

plt.show()

# เซฟเป็นไฟล์ Excel ไปไว้ในโฟลเดอร์ revenue ใน Drive
path_to_save = '/content/drive/MyDrive/12. Banyasookjai/10 อันดับสินค้าที่มี % กำไรขั้นต้นสูงที่สุด (และมียอดขายเกิน 20 ชิ้น).xlsx'

# คำสั่งเซฟ (index=False คือไม่เอาตัวเลขลำดับแถว 0,1,2,3... ติดไปด้วย)
top_margin_products.to_excel(path_to_save, index=False)
print("✅ เซฟไฟล์ Excel สำเร็จแล้ว ไปดูใน Drive ได้เลย!")
```

#Insight ข้อที่ 1: Market Basket Analysis (การหาว่าสินค้าอะไรที่ลูกค้ามักซื้อคู่กัน)

```python
# ---------------------------------------------------------
# 🤝 Insight 1: Market Basket Analysis (ฉบับประหยัด RAM + เร็วสุดๆ)
# ---------------------------------------------------------
import pandas as pd
# 🌟 เปลี่ยนมาใช้ fpgrowth แทน apriori (เร็วกว่าและกินแรมน้อยกว่ามาก)
from mlxtend.frequent_patterns import fpgrowth
from mlxtend.frequent_patterns import association_rules

print("✂️ 1. กำลังกรองข้อมูลเพื่อประหยัด RAM...")

# เทคนิคประหยัดแรม: กรองเอาเฉพาะ "บิลที่มีการซื้อสินค้า 2 ชิ้นขึ้นไป" เท่านั้น
# (เพราะบิลชิ้นเดียวหาของคู่กันไม่ได้อยู่แล้ว)
item_counts = df_clean.groupby('เลขที่ใบเสร็จรับเงิน')['รายการสินค้า'].nunique()
multi_item_bills = item_counts[item_counts > 1].index
df_filtered = df_clean[df_clean['เลขที่ใบเสร็จรับเงิน'].isin(multi_item_bills)]

print("⏳ 2. กำลังสร้างตารางตะกร้าสินค้า...")
basket = df_filtered.groupby(['เลขที่ใบเสร็จรับเงิน', 'รายการสินค้า'])['จำนวน'] \
                 .sum().unstack().reset_index().fillna(0).set_index('เลขที่ใบเสร็จรับเงิน')

print("⚡ 3. กำลังแปลงข้อมูลเป็น True/False (ใช้พื้นที่แรมน้อยที่สุด)...")
# แปลงเป็น True/False ซึ่งไลบรารี mlxtend ชอบมากและกินแรมน้อยสุด
basket_sets = basket.astype(bool)

print("🧠 4. กำลังค้นหากฎคู่สินค้าด้วยอัลกอริทึม FP-Growth (เร็วกว่าเดิม 10 เท่า!)...")
# ใช้ fpgrowth แทน apriori
frequent_itemsets = fpgrowth(basket_sets, min_support=0.005, use_colnames=True)

if not frequent_itemsets.empty:
    rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1.0)

    # กำจัดตัวซ้ำสลับหัวท้าย
    rules = rules.sort_values('confidence', ascending=False)
    rules['คู่สินค้ารวม'] = rules.apply(lambda row: frozenset(row['antecedents']).union(frozenset(row['consequents'])), axis=1)
    rules_unique = rules.drop_duplicates(subset=['คู่สินค้ารวม']).copy()

    # จัดตารางสวยๆ
    rules_clean = pd.DataFrame()
    rules_clean['สินค้าตัวหลัก'] = rules_unique['antecedents'].apply(lambda x: ', '.join(list(x)))
    rules_clean['สินค้าที่หยิบซื้อคู่กัน'] = rules_unique['consequents'].apply(lambda x: ', '.join(list(x)))
    rules_clean['โอกาสซื้อคู่กัน (%)'] = (rules_unique['confidence'] * 100).round(2)
    rules_clean['ความเชื่อมโยง (Lift > 1)'] = rules_unique['lift'].round(2)

    print("\n🎉 วิเคราะห์เสร็จสิ้น! นี่คือ 10 คู่สินค้าเน้นๆ ที่ลูกค้านิยมซื้อคู่กัน:")
    display(rules_clean.head(5))

    # เซฟลง Drive
    path_to_save = '/content/drive/MyDrive/12. Banyasookjai/revenue/กฎคู่สินค้า_MarketBasket.xlsx'
    rules_clean.to_excel(path_to_save, index=False)
    print("\n✅ เซฟไฟล์ Excel ลง Drive สำเร็จครับ!")

else:
    print("\n❌ ไม่พบสินค้าที่ซื้อคู่กัน ลองปรับ min_support ให้น้อยลงอีกครับ")
```

#ข้อที่ 6: วิเคราะห์ความคุ้มค่าของการให้ส่วนลด

```python
# ---------------------------------------------------------
# 🏷️ Insight 6: วิเคราะห์ความคุ้มค่าของการให้ส่วนลด
# ---------------------------------------------------------
import matplotlib.pyplot as plt
import seaborn as sns

# 1. จัดกลุ่มข้อมูลเป็นรายใบเสร็จ (ยุบรวมสินค้าทุกชิ้นในบิลเดียวกัน)
discount_analysis = df_clean.groupby('เลขที่ใบเสร็จรับเงิน').agg({
    'ราคาขายรวม': 'sum',  # ยอดเงินรวมต่อบิล
    'จำนวน': 'sum',        # จำนวนชิ้นรวมต่อบิล
    'ส่วนลดรวม': 'sum',    # ส่วนลดรวมต่อบิล
    'กำไรรวม': 'sum'       # กำไรรวมต่อบิล
}).reset_index()

# 2. สร้างคอลัมน์ใหม่เพื่อแปะป้ายว่าบิลนี้ "มีส่วนลด" หรือ "ไม่มีส่วนลด"
# โดยเช็คว่าถ้า 'ส่วนลดรวม' มากกว่า 0 แปลว่าบิลนั้นมีการให้ส่วนลด
discount_analysis['กลุ่มส่วนลด'] = discount_analysis['ส่วนลดรวม'].apply(
    lambda x: 'ให้ส่วนลด' if x > 0 else 'ไม่ให้ส่วนลด'
)

# 3. หาค่าเฉลี่ยเปรียบเทียบระหว่าง 2 กลุ่ม
discount_summary = discount_analysis.groupby('กลุ่มส่วนลด').agg({
    'เลขที่ใบเสร็จรับเงิน': 'count', # นับจำนวนบิลทั้งหมด
    'ราคาขายรวม': 'mean',            # ยอดใช้จ่ายเฉลี่ยต่อบิล (Ticket Size)
    'จำนวน': 'mean',                 # จำนวนชิ้นเฉลี่ยต่อบิล (Basket Size)
    'กำไรรวม': 'mean'                # กำไรเฉลี่ยต่อบิล
}).reset_index()

# เปลี่ยนชื่อคอลัมน์ให้แสดงผลภาษาไทยอ่านง่ายๆ
discount_summary.rename(columns={
    'เลขที่ใบเสร็จรับเงิน': 'จำนวนบิลทั้งหมด (ใบ)',
    'ราคาขายรวม': 'ยอดซื้อเฉลี่ย (บาท/บิล)',
    'จำนวน': 'จำนวนชิ้นเฉลี่ย (ชิ้น/บิล)',
    'กำไรรวม': 'กำไรเฉลี่ย (บาท/บิล)'
}, inplace=True)

print("🏷️ เปรียบเทียบพฤติกรรม: บิลที่ให้ส่วนลด VS บิลที่ไม่ให้ส่วนลด")
display(discount_summary.round(2)) # .round(2) คือปัดทศนิยม 2 ตำแหน่ง

# 4. วาดกราฟเปรียบเทียบ "ยอดซื้อเฉลี่ยต่อบิล" ให้เห็นภาพชัดเจน
plt.figure(figsize=(8, 6))
# ใช้กราฟแท่ง (Barplot) เปรียบเทียบ 2 กลุ่ม
sns.barplot(data=discount_summary, x='กลุ่มส่วนลด', y='ยอดซื้อเฉลี่ย (บาท/บิล)', palette='pastel')

plt.title('🏷️ เปรียบเทียบยอดซื้อเฉลี่ย: ให้ส่วนลด VS ไม่ให้ส่วนลด')
plt.ylabel('ยอดใช้จ่ายเฉลี่ยต่อบิล (บาท)')
plt.xlabel('')

# 5. เขียนตัวเลขกำกับไว้บนยอดกราฟแท่ง
for index, value in enumerate(discount_summary['ยอดซื้อเฉลี่ย (บาท/บิล)']):
    plt.text(index, value + (value * 0.02), f'{value:,.2f} ฿', ha='center', fontsize=14)

plt.show()
```
